```python
import pandas as pd
import plotly as pl
```

```python
df_AAPL = pd.read_csv('AAPL_stock_price.csv')
df_MSFT = pd.read_csv('MSFT_stock_price.csv')
df_AAPL.dropna(axis=1)
df_MSFT.dropna(axis=1)
```

```python
import plotly.graph_objs as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_AAPL['Date'], y=df_AAPL['Close Price'],
                         mode='lines', name='AAPL',
                         line=dict(color='blue')))
fig.add_trace(go.Scatter(x=df_MSFT['Date'], y=df_MSFT['Close Price'],
                         mode='lines', name='MSFT',
                         line=dict(color='red')))
fig.update_layout(title='PreÃ§o de Fechamento da AAPL e MSFT',
                  xaxis_title='Data',
                  yaxis_title='PreÃ§o de Fechamento',
                  shapes=[
                      dict(
                          x0='2019-01-22', x1='2019-01-22',
                          y0=0, y1=1,
                          xref='x', yref='paper',
                          line_width=2, line_color='blue'
                      ),dict(
                          x0='2012-09-25', x1='2012-09-25',
                          y0=0, y1=1,
                          xref='x', yref='paper',
                          line_width=2, line_color='blue'
                      ),dict(
                          x0='2020-11-25', x1='2020-11-25',
                          y0=0, y1=1,
                          xref='x', yref='paper',
                          line_width=2, line_color='red'
                      ),dict(
                        type='line',
                        x0=df_AAPL['Date'].min(),  
                        y0=210,  
                        x1=df_AAPL['Date'].max(), 
                        y1=210,  
                        line=dict(
                            color='green',  
                            width=2,  
                            dash='dash'  
                        )
                    )
                  ],
                  annotations=[dict(
                      x='2019-01-22', y=0.95, yref='paper',
                      showarrow=False,
                      xanchor='right',
                      text='Increase Period'
                  ),dict(
                      x='2012-09-25', y=0.98, yref='paper',
                      showarrow=False,
                      xanchor='right',
                      text='Decrease Period'
                  ),dict(
                      x='2020-11-25', y=0.98, yref='paper',
                      showarrow=False,
                      xanchor='left',
                      text='Increase Period'
                  )]
                 )
fig.show()
```
